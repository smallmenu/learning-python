```python
import matplotlib.pyplot as plt
import numpy as np

# 预设样式
print(plt.style.available)
plt.style.use('seaborn-v0_8')

# 绘图尺寸
# plt.figure(figsize=(12, 2))

# linspace 用于生成等间距的数组，
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# 绘制多条线
fig, ax = plt.subplots(figsize=(12, 3))

# 绘制十字坐标轴
ax.axhline(0, color='gray', lw=1)
ax.axvline(0, color='gray', lw=1)

ax.plot(x, y1, 'b-', label='Sin(x)')
ax.plot(x, y2, 'r-', label='Cos(x)')

# 改变X轴的刻度
x_ticks = np.linspace(-2 * np.pi, 2 * np.pi, 17)
x_ticklabels = ['-2π', '-1*3π/4', '-1π/2', '-1*π/4', '-π', '-3π/4', '-π/2', '-π/4', 0, 'π/4', 'π/2', '3π/4', 'π', '1*π/4', '1*π/2', '1*3π/4', '2π']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Multiple Lines')

# 显示标签
plt.legend()
plt.show()
```
