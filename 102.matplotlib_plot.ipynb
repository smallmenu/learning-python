```python
import matplotlib.pyplot as plt
import numpy as np

# linspace 用于生成等间距的数组，
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# 绘图尺寸
plt.figure(figsize=(12, 2))

# subplot 函数用于在一个已有的画布上创建一个子图，通过指定子图的行数、列数以及子图的位置来确定子图的位置
# plot 用于绘制折线图、散点图散点图、条形图、面积图
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Sin(x)')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Cos(x)')

# 绘制图形
plt.show()
```

```python


# 使用subplots()函数用于创建多个子图，每个子图都是一个Axes对象，这是官方推荐的方式
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))
axes[0].plot(x, y1)
axes[0].set_title('Sinx(x)')
axes[1].plot(x, y2)
axes[1].set_title('Cos(x)')

plt.show()
```

```python
# 绘制多条线
fig, ax = plt.subplots(figsize=(12, 2))

ax.plot(x, y1, '', label='Sin(x)')
ax.plot(x, y2, 'r-', label='Cos(x)')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Multiple Lines')

# 显示标签
plt.legend()
plt.show()
```

```python
# 绘制多条线
fig, ax = plt.subplots(figsize=(12, 4))

# 绘制十字坐标轴
ax.axhline(0, color='gray', lw=1)
ax.axvline(0, color='gray', lw=1)

# 设置线条样式
ax.plot(x, y1, '', label='Sin(x)')
ax.plot(x, y2, 'r-.', label='Cos(x)')

# 绘制网格
ax.grid(True)

# XY轴的长度范围
ax.axis([-10, 10, -1, 1])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Multiple Lines')

# 显示标签
plt.legend()
plt.show()
```
